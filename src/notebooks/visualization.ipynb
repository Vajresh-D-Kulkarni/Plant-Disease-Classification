```python
from pathlib import Path

history_path = PROJECT_ROOT / "experiments" / "baseline_cnn" / "baseline_history.json"
output_dir = PROJECT_ROOT / "experiments" / "baseline_cnn" / "visualizations"

try:
    plot_training_curves(history_path, output_dir)
    print("Training curves plotted successfully")
except Exception as e:
    print(f"Error plotting training curves: {e}")
```

```python
from src.data.data_loader import load_datasets
try:
    _, val_ds, class_names = load_datasets(base_path=PROJECT_ROOT / "dataset" / "New Plant Diseases Dataset(Augmented)")
except Exception as e:
    print(f"Error loading datasets: {e}")
```

```python
model_path = PROJECT_ROOT / "models" / "baseline_cnn.h5"
try:
    from tensorflow import keras
    model = keras.models.load_model(model_path)
    print("Model loaded successfully")
except Exception as e:
    print(f"Error loading model: {e}")
```

```python
import numpy as np

y_true = []
y_pred = []

for images, labels in val_ds:
    preds = model.predict(images, verbose=1)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

print("y_true shape:", y_true.shape)
print("y_pred shape:", y_pred.shape)
```

```python
from src.visualization.metrics import compute_and_plot_metrics
from src.visualization.confusion_matrix import plot_confusion_matrix
```

```python
try:
    metrics = compute_and_plot_metrics(y_true=y_true, y_pred=y_pred, class_names=class_names, output_dir=output_dir)
except Exception as e:
    print(f"Error computing metrics: {e}")
```

```python
try:
    plot_confusion_matrix(y_true=y_true, y_pred=y_pred, class_names=class_names, output_dir=output_dir)
except Exception as e:
    print(f"Error plotting confusion matrix: {e}")
```
