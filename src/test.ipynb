```python
import json
import numpy as np
import tensorflow as tf
from pathlib import Path
from PIL import Image
from tensorflow.keras.applications.efficientnet import preprocess_input
```

```python
PROJECT_ROOT = Path.cwd().parent

MODEL_PATH = PROJECT_ROOT / "models" / "efficientnet_b3_stage2.keras"
CLASS_NAMES_PATH = PROJECT_ROOT / "models" / "class_names.json"

TEST_IMAGE = (
    PROJECT_ROOT
    / "dataset"
    / "New Plant Diseases Dataset(Augmented)"
    / "test"
    / "PotatoEarlyBlight3.JPG"
)

MODEL_PATH.exists(), CLASS_NAMES_PATH.exists(), TEST_IMAGE.exists()
```

```python
model = tf.keras.models.load_model(MODEL_PATH)

with open(CLASS_NAMES_PATH) as f:
    class_names = json.load(f)

print("Total classes:", len(class_names))
print("First 5 classes:", class_names[:5])
```

```python
img = Image.open(TEST_IMAGE).convert("RGB")
img = img.resize((224, 224))

img_array = np.array(img)
img_array = preprocess_input(img_array)
img_array = np.expand_dims(img_array, axis=0)
```

```python
preds = model.predict(img_array)

pred_idx = int(np.argmax(preds))
confidence = float(np.max(preds))

print("Predicted class:", class_names[pred_idx])
print("Confidence:", confidence)
```

```python
top5 = np.argsort(preds[0])[::-1][:5]

print("\nTop-5 predictions:")
for idx in top5:
    print(f"{class_names[idx]:45s}  {preds[0][idx]:.4f}")
```
